# RFM ANALYSIS

FM (Recency, Frequency, Monetary) analysis is a behavior-based approach grouping customers into segments. 
It groups the customers on the basis of their previous purchase transactions. How recently, how often, and how much did a 
customer buy. RFM filters customers into various groups for the purpose of better service. It helps managers to identify 
potential customers to do more profitable business.

Recency (R): Who have purchased recently? Number of days since last purchase (least recency)

Frequency (F): Who has purchased frequently? It means the total number of purchases. ( high frequency)

Monetary Value(M): Who have high purchase amount? It means the total money customer spent (high monetary value)

```python
# Recency = (Latest_date - Date) basis individual customer id
# Frequency = Total no of Count(Invoice No) basis individual customer id
# Monetory = Total sum of price(basis individual customer id)
```

```python
RFMscores = data_new.groupby('CustomerID').agg({'Date' : lambda x : (Latest_date - x.max()).days,
                                                'InvoiceNo' : lambda x : x.count(),
                                                'Price' : lambda x : x.sum()})
```

```python
RFMscores.rename(columns = {'Date': 'Recency', 'InvoiceNo':'frequency', 'Price':'Monetory'}).reset_index()
```

```python
RFMscores.Recency.describe()
```

```python
RFMscores.frequency.describe()
```

```python
RFMscores.Monetory.describe()
```

### quantile percentage


```python
quantile = RFMscores.quantile(q = [0.25,0.5,0.75])
quantile = quantile.to_dict()
quantile
```

```python
def RScore(x,p,d):
    if x<=d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4
    
def FnMScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1
    
```

```python
RFMscores['R'] = RFMscores['Recency'].apply(RScore, args = ('Recency',quantile))
RFMscores['F'] = RFMscores['frequency'].apply(FnMScore, args = ('frequency',quantile))
RFMscores['M'] = RFMscores['Monetory'].apply(FnMScore, args = ('Monetory',quantile))
```

```python
RFMscores
```

```python
RFMscores['RFMGroups'] = RFMscores['R'].map(str)+ RFMscores['F'].map(str) + RFMscores['M'].map(str)
RFMscores
```

```python
RFMscores['RFMSum'] = RFMscores[['R','F','M']].sum(axis = 1)
RFMscores.reset_index()
```

```python
RFMscores['RFMSum'].value_counts()
```

### exporting the csv files


```python
RFMscores.to_csv('Rfm_analysis.csv')
```

```python
Loyality_level = ['Gold','platinum','silver','Titanium']
score_cuts = pd.qcut(RFMscores['RFMSum'],q = 4,labels = Loyality_level)
score_cuts
```

```python
RFMscores['Loyality_level'] = score_cuts.values
RFMscores.reset_index()
RFMscores
```

### plotting pie chart to display segmentation based on RFM


```python
plt.figure(figsize=(10,10))
RFMscores['Loyality_level'].value_counts().plot(kind='pie',autopct='%.1f')
plt.title("Customer Segmentation",size=15)
plt.axis('equal') 
plt.show()
```
